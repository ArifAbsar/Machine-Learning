<h3>Import necessary libraries</h3>

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

<h3>Load CSV data file </h3>

```python
df=pd.read_csv('weatherAUS.csv')
df.head()
```

<h3>DATA PROCESSING </h3>

First we need to change categorical value to numeric value and for we will use <b>ONE-HOT ENCODING </b>

```python
processed=pd.get_dummies(data=df,columns=['Location','RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
processed
```

Feature 'RainTommorow' is in categorical and we will switch it to [0,1]

```python
processed.replace(['No','Yes'],[0,1],inplace=True)
```

<h4>Now we drop the 'Date' feature and set out remaining feature to X and y for train/test split</h4>

```python

processed.drop('Date',axis=1,inplace=True)
```

```python

#processed = processed.astype(float)
data_cleaned = processed[~np.isnan(processed).any(axis=1)]
data_cleaned
```

```python
X=data_cleaned.drop(columns='RainTomorrow',axis=1)
y = data_cleaned['RainTomorrow']
```

<h4>Train/test split</h4>

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

<h2>LINEAR REGRESSION</h2>

```python
Linearreg=LinearRegression()
Linearreg.fit(X_train,y_train)
```

```python
y_hat=Linearreg.predict(X_test)
print(y_hat)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

LinearRegression_MAE = mean_absolute_error(y_test, y_hat)
LinearRegression_MSE = mean_squared_error(y_test, y_hat)
LinearRegression_R2 = r2_score(y_test, y_hat)

print("MAE = ",LinearRegression_MAE)
print("MSE = ",LinearRegression_MSE)
print("R2_SCORE = ",LinearRegression_R2)
```

<h3>DECISION TREE</h3>

```python
rainTree=DecisionTreeClassifier(criterion='entropy',max_depth=4)
rainTree
rainTree.fit(X_train,y_train)
pred=rainTree.predict(X_test)
from sklearn.metrics import accuracy_score,jaccard_score,f1_score
print("Accuracy = ",accuracy_score(y_test,pred))
print("Jacard Index = ",jaccard_score(y_test,pred,pos_label=0))
print("F1 Score = ",f1_score(y_test, pred, average='weighted'))
```

<h3>SVM</h3>

```python
model=svm.SVC(kernel='rbf')
model.fit(X_train,y_train)
s=model.predict(X_test)
print("Accuracy = ",accuracy_score(y_test,s))
print("Jacard Index = ",jaccard_score(y_test,s,pos_label=0))
print("F1 Score = ",f1_score(y_test, s, average='weighted'))
```

<h3>KNN</h3>

```python
k=4
KNN=KNeighborsClassifier(n_neighbors=k).fit(X_train,y_train)
prediction=KNN.predict(X_test)
print("Accuracy = ",accuracy_score(y_test,prediction))
print("Jacard Index = ",jaccard_score(y_test,prediction,pos_label=0))
print("F1 Score = ",f1_score(y_test, prediction, average='weighted'))
```

<h3>Logistic Regression</h3>

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
Logistic=LogisticRegression(C=0.01,solver='liblinear').fit(X_train,y_train)
yhat=Logistic.predict(X_test)
print("Accuracy = ",accuracy_score(y_test,yhat))
print("Jacard Index = ",jaccard_score(y_test,yhat,pos_label=0))
print("F1 Score = ",f1_score(y_test,yhat, average='weighted'))
print("Log loss = ",log_loss(y_test,yhat))
```
